**9. Using matplotlib plot the following functions on the domain: x = 0.5 to 100.0 in steps of 0.5.**



```python
#import matplotlib and numpy
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#define all the function (index same as in the question)

def func1(x):
    return x

def func2(x):
    return x**2

def func3(x):
    return x**3/100

def func4(x):
    return np.sin(x)

def func5(x):
    return np.sin(x)/x

def func6(x):
    return np.log(x)

def func7(x):
    return np.exp(x)

```

```python
#define x
x = np.arange(0.5, 100.0, 0.5)

# create a plot figure
plt.figure(figsize=(10,20)) 
# create the first of two panels and set current axis
plt.subplot(7, 1, 1)
plt.plot(x, func1(x),label='y = x')

# create the second panel 
plt.subplot(7, 1, 2)
plt.plot(x, func2(x));

# create the third panel 
plt.subplot(7, 1, 3)
plt.plot(x, func3(x));

# create the fourth panel 
plt.subplot(7, 1, 4)
plt.plot(x, func4(x));

# create the fifth panel 
plt.subplot(7, 1, 5)
plt.plot(x, func5(x));

# create the sixth panel 
plt.subplot(7, 1, 6)
plt.plot(x, func6(x));

# create the seventh panel 
plt.subplot(7, 1, 7)
plt.plot(x, func7(x));
```
